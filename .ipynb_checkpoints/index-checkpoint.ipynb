```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel('flight_price.xlsx')
df1.isnull().sum()
```

```python
df1['Date'] = df1['Date_of_Journey'].str.split('/').str[0].astype(int)
df1['Month'] = df1['Date_of_Journey'].str.split('/').str[1].astype(int)
df1['Year'] = df1['Date_of_Journey'].str.split('/').str[2].astype(int)
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.info()
```

```python
df1.drop(['Route','Date_of_Journey'],axis=1,inplace=True)
df1.head()
```

```python
df1['Dep_hr'] = df1['Dep_Time'].str.split(':').str[0].astype(int)
df1['Dep_min'] = df1['Dep_Time'].str.split(':').str[1].astype(int)
df1.head()
```

```python
df1['Arrival_Time'] = df1['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df1.head()
```

```python
df1['Arrival_hr'] = df1['Arrival_Time'].str.split(':').str[0].astype(int)
df1['Arrival_min'] = df1['Arrival_Time'].str.split(':').str[1].astype(int)
df1.head()
```

```python
df1.info()
```

```python
df1.drop(['Dep_Time','Arrival_Time'],axis=1,inplace=True)
df1.info()
```

```python
df1['Duration_hr'] = df1['Duration'].str.split(' ').str[0].str[:-1]
df1.head()
```

```python
df1['Duration_hr'] = df1['Duration_hr'].astype(int)
df1['Duration_min'] = df1['Duration'].str.split(' ').str[0].str[:-1].astype(int)
df1.info()
```

```python
df1.drop('Duration',axis=1,inplace=True)
```
